```python
main_df=df[(~df['Code'].isnull()) & (df['Entity']!='World')]
main_df
```

```python
main_df['Total']=main_df[main_df.columns[3:]].sum(axis=1)
```

```python
## this could also work
df['Total']=df.iloc[:,3:].sum(axis=1)
```

```python
top5=main_df.groupby('Entity')['Total'].sum().nlargest().index
```

```python
df_top5=main_df[main_df['Entity'].isin(top5)]
```

```python
df_top5
```

```python
sns.lineplot(df_top5,x='Year',y='Total',hue='Entity')
```

```python
sum_ng=[]
index=[]
for i in df_nig.columns[3:-1]:
    ng=df_nig[i].sum()
    sum_ng.append(ng)
    index.append(i)
    trial=pd.DataFrame(sum_ng,index=index,columns=['Total'])
```

```python
nig_top5=trial.sum(axis=1).nlargest().index
```

```python
df_nig_top5=df_nig[['Year',*nig_top5]]
```

```python
df_nig_top5.columns
```

```python
for i in df_nig_top5.columns[1:]:
    sns.lineplot(df_nig_top5,x='Year',y=i,label=i)
    plt.ylabel('Annual DEaths')
```
